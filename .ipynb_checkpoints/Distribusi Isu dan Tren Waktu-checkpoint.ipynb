```python
import glob
import json
import pandas as pd

files = glob.glob("dataset_280426/selectedanswer*.json")

# Baca semua file dan kumpulkan dalam satu list
list_data = []
for f in files:
    with open(f, encoding='utf-8-sig') as data_file:
        list_data.append(json.load(data_file))

# Gabung jadi satu tabel besar
df = pd.DataFrame(list_data)

df.head()
```

```python
data_split = df.explode('data')
data_normalized = pd.json_normalize(data_split['data'])
df_tren = data_normalized
display(data_normalized.head())
```

### Top 10 Kategori Pertanyaan Tiap Bulan (persentase)

```python
# 1. Cari 10 kategori terbanyak
top_10_categories = df_tren1['category'].value_counts().nlargest(10).index

# 2. Tambahkan .copy() di akhir filter agar tidak kena SettingWithCopyWarning atau Indexing Error
df_top10 = df_tren1[df_tren1['category'].isin(top_10_categories)].copy()

# Reset index untuk memastikan tidak ada label yang bentrok/duplikat
df_top10 = df_top10.reset_index(drop=True)

# 3. Buat kolom bulan_tahun menggunakan .loc agar aman
df_top10['bulan_tahun'] = df_top10['question_date'].dt.to_period('M')

# Karena crosstab kadang sensitif dengan tipe data 'Period', 
# kita ubah ke string agar visualisasi heatmap di sumbu X jadi rapi
df_top10['bulan_tahun'] = df_top10['bulan_tahun'].astype(str)

tren_top10 = pd.crosstab(df_top10['category'], df_top10['bulan_tahun'])

# --- LANGKAH REVISI UNTUK PERSENTASE ---
# Menghitung persentase per bulan (axis=0 artinya total per kolom/bulan adalah 100%)
tren_persen = tren_top10.div(tren_top10.sum(axis=0), axis=1)

# 5. Visualisasi Heatmap
plt.figure(figsize=(15, 8))
sns.heatmap(tren_persen, 
            annot=True,          
            fmt='.1%',             # Mengubah angka desimal menjadi format persen (misal: 0.123 -> 12.3%)
            cmap='YlGnBu',       
            linewidths=.5,
            cbar_kws={'label': 'Proporsi Keluhan per Bulan'})

plt.title('Proporsi Tren 10 Kategori Pertanyaan Terbanyak Per Bulan', fontsize=15)
plt.xlabel('Periode', fontsize=12)
plt.ylabel('Kategori', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

### Persebaran Harian Tren Kategori Bulan Januari

```python
# 1. Pastikan kolom tanggal adalah datetime
df_tren1['question_date'] = pd.to_datetime(df_tren1['question_date'])

# 2. Ambil Top 5 Kategori dari seluruh data
top_5_categories = df_tren1['category'].value_counts().nlargest(5).index

# 3. Filter data Januari 2026 DAN masuk dalam Top 5
mask = (df_tren1['question_date'].dt.year == 2026) & \
       (df_tren1['question_date'].dt.month == 1) & \
       (df_tren1['category'].isin(top_5_categories))

df_jan_filtered = df_tren1[mask].copy()

# 4. Agregasi data: Hitung jumlah pertanyaan per hari per kategori
# Kita buat tabel pivot: Baris = Tanggal, Kolom = Kategori, Value = Count
df_pivot = df_jan_filtered.groupby([df_jan_filtered['question_date'].dt.date, 'category']).size().unstack(fill_value=0)

# 5. Plotting dari tabel pivot
ax = df_pivot.plot(figsize=(15, 6), marker='o', markersize=4)

plt.title('Tren Harian Top 5 Kategori di Bulan Januari 2026', fontsize=14)
plt.ylabel('Jumlah Pertanyaan')
plt.xlabel('Tanggal')
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 1. Tentukan kategori pilihanmu sendiri dalam bentuk list
# Pastikan tulisannya sama persis (case-sensitive) dengan yang ada di kolom 'category'
selected_categories = ['Pendaftaran dan Aktivasi Wajib Pajak'] 

# 2. Filter data Januari 2026 DAN masuk dalam kategori pilihan tadi
mask = (df_tren1['question_date'].dt.year == 2026) & \
       (df_tren1['question_date'].dt.month == 1) & \
       (df_tren1['category'].isin(selected_categories))

df_jan_filtered = df_tren1[mask].copy()

# 3. Lanjutkan ke tahap pivot dan plotting seperti sebelumnya
df_pivot = df_jan_filtered.groupby([df_jan_filtered['question_date'].dt.date, 'category']).size().unstack(fill_value=0)
df_pivot.index = df_pivot.index.astype(str)

fig, ax = plt.subplots(figsize=(15,6))
df_pivot.plot(kind='line', marker='o', markersize=4, ax=ax)

plt.xticks(
    ticks=range(len(df_pivot)), 
    labels=df_pivot.index, 
    rotation=45, 
    ha='right'
)
# ---------------------------

plt.title('Tren Harian Bulan Januari Kategori: Pendaftaran dan Aktivasi Wajib Pajak', fontsize=14)
plt.ylabel('Jumlah Pertanyaan')
plt.xlabel('Tanggal')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
